```
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures 
import matplotlib.pyplot as plt 
import numpy as np

 
def polyfit2(x,y,n):
    lin_reg = LinearRegression()
    poly_reg = PolynomialFeatures(degree=n)

    def inv(A):
      return np.linalg.inv(A) #Matriz inversa de la matriz A
    def trans(A):
      return A.getT()  #trnspuesta de la matriz A
    def prod(A,B):
      return np.dot(A,B) #producto entre los vectores A y B

    xlen = len(x) #calcula la longitud del vector x
    ylen = len(y)
    one = np.ones((xlen,n+1),dtype=int) #matriz de 1's de dimensiones (xlen, n+1)
    c1=one[:,[1]] #seleccionamos la columna 1 de la matriz one
    xT=np.matrix(x) #Convierte los datos a una matriz de numpy
    yT=np.matrix(y)
    c2=xT.getT()
    c3=np.power(c2,2) #eleva al cuadrado los datos de la matriz c2
    c4=np.power(c3,2) #eleva al cuadrado los datos de la matriz c3
    c5=np.power(c4,2) #eleva al cuadrado los datos de la matriz c4
    c6=np.power(c5,2) #eleva al cuadrado los datos de la matriz c5
    A1=np.hstack([c1,c2]) #agrupa las columnas c1, c2 en una matriz A  (Grado 1)
    A2=np.hstack([c1,c2,c3]) #agrupa las columnas c1, c2 y c3 en una matriz A  (Grado 2)
    A3=np.hstack([c1,c2,c3,c4]) #agrupa las columnas c1, c2,c3,c4 en una matriz A  (Grado 3)
    A4=np.hstack([c1,c2,c3,c4,c5]) #agrupa las columnas c1, c2, c3, c4, c5 en una matriz A  (Grado 4)
    A5=np.hstack([c1,c2,c3,c4,c5,c6]) #agrupa las columnas c1, c2, c3, c4, c5 y c6 en una matriz A  (Grado 5)
    

    ajuste1=prod(prod(inv(prod(trans(A1),A1)),trans(A1)),trans(yT)) #la funcion prod realiza los productos de las matrices que recibe como argumentos
    print(ajuste1)
    
    #grafica 1
    plt.scatter(x,y)
    x1 = np.arange(-5,8,0.1)
    y1 = 3.57 - 0.88*x1 
    plt.plot(x1,y1)
    plt.show()
    
    ajuste2=prod(prod(inv(prod(trans(A2),A2)),trans(A2)),trans(yT)) #la funcion prod realiza los productos de las matrices que recibe como argumentos
    print(ajuste2)
    #grafica 2
    
    plt.scatter(x,y)
    x2 = np.arange(-5,8,0.1)
    y2 = -0.0378*np.power(x2,2) - 0.8106*x2 + 3.75
    plt.plot(x2,y2)
    plt.show()


    ajuste3=prod(prod(inv(prod(trans(A3),A3)),trans(A3)),trans(yT)) #la funcion prod realiza los productos de las matrices que recibe como argumentos
    print(ajuste3)
    #grafica 3
    
    plt.scatter(x,y)
    x3 = np.arange(-5,8,0.1)
    y3 = 0.0537037*np.power(x3,3) -0.91666667*np.power(x3,2) + -0.98148148*x3 + 5.84444444
    plt.plot(x3,y3)
    plt.show()


    #ajuste4=prod(prod(inv(prod(trans(A4),A4)),trans(A4)),trans(yT)) #la funcion prod realiza los productos de las matrices que recibe como argumentos
    #print(ajuste4)
   
data = {'x':[1.0,-2.0,3.0,4.0],
'y':[4.0,5.0,-1.0,1.0]

}
polyfit2(data['x'],data['y'],2) #el resultado es el polinomio y=-0.0378x^2 - 0.8106x + 3.75


    

```
